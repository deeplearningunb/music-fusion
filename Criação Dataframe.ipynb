## Criação Dataframe

Consiste na criação de um dataset de musicas por meio de um dataset de metadados com **_Track name_**  e **_Artist name_**. Com esses dados serão feitas pesquisas na api do vagalume https://api.vagalume.com.br/. Caso não seja encontrado deve-se adicionar manualmente novas musicas.

Um passo do préprocessameto dos dados já são feitas nessa fase (transformação do texto para letras minúsculas)

```python
import requests
import json
import pandas as pd

def generate_dataset(metadata_path='metadata_musics.csv', dataset_path='musics.csv'):
    key = '53f2e9702783aead0ec93fe8be2333df'

    dt_mus = pd.read_csv(metadata_path, encoding = "utf-8")
    dt_mus.head()

    list_musics = []
    for index, row in dt_mus.iterrows():
        art = row[' Artist name']
        mus = row['Track name']

        response = requests.get(
            f'https://api.vagalume.com.br/search.php?art={art}&mus={mus}&apikey={key}'
        )
        if response.status_code == 200:
            data = response.content.decode('utf-8')

            dict_json = json.loads(data)
            if dict_json['type'] != 'notfound' and dict_json['type'] != 'song_notfound':
                if len(dict_json['mus']) > 0:
                    dict_json['mus'][0]['text']
                    list_musics.append(
                        {'artist': art,
                        'track': mus,
                        'music': dict_json['mus'][0]['text'].lower()}
                    )

    dframe = pd.DataFrame.from_records(list_musics)
    dframe.to_csv(dataset_path, encoding='utf-8', index=False)
    return dframe
```

```python
df_frame = generate_dataset()
```
